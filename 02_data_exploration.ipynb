```python
print(df_train.shape)
df_train = df_train.dropna()
print(df_train.shape)
```

```python
# Input
# - cell_type: categorical (5)
# - sm_name: categorical (144)
# - gene_name: categorical (18,000)
# - controol: boolean
# Output
# - differential gene expression: real number
#
# Representation:
# - We need to encode some of the inputs with dense vectors, given the cardinality
#   of the gene_name input, we can't use sparse one-hot vectors (aka dummy variables)
#   because it would take too much memory.
#
# Model architecture
# The simplest way to encode the inputs is with a trainable embedding layer.
# The output is a real number (the model is a regressor) so the simplest possible model
# has a single-unit output layer with linear activation.
```

```python
# sparse encoding
def encode(idx, size):
  x = [0] * size
  x[idx] = 1
  return x

encode(4, 6)
```

```python
# indexing
create_index = lambda names: {name:idx for idx, name in enumerate(sorted(set(names)))}

def get_indexing(names, index):
  index = [index[name] for name in names]
  return index

create_index(df_train['cell_type'].values)
```

```python
# create indexing tables
gene_names = list(df_train.columns[5:])
indices = {
    'cell_type': create_index(df_train['cell_type'].values),
    'sm_name': create_index(df_train['sm_name'].values),
    'gene_name': create_index(gene_names),
}
list(indices['gene_name'].items())[:10]
```

```python
# populate the feature matrix and label array
feature_matrix_X = []
label_Y = []

# convert from dataframe to lists
cell_types = df_train['cell_type'].values
sm_names = df_train['sm_name'].values
controls = df_train['control'].values
ge_data = df_train[gene_names].values

# cell-sm combination
for cell_type, sm_name, control, gene_expressions in zip(cell_types, sm_names, controls, ge_data):
  # if not control:
  #   continue
  # print(cell_type, sm_name, control)
  # indexing of row values
  cell_type_id = indices['cell_type'][cell_type]
  sm_name_id = indices['sm_name'][sm_name]
  control_id = int(control)

  # ge columns
  for gene_name, gene_expression in zip(gene_names, gene_expressions):
    # print(gene_name, gene_expression)
    gene_name_id = indices['gene_name'][gene_name]
    feature_matrix_X.append((cell_type_id, sm_name_id, control_id, gene_name_id))
    label_Y.append(gene_expression)
  #   break
  # break

feature_matrix_X = np.array(feature_matrix_X, dtype='int32').reshape(-1,4)
label_Y = np.array(label_Y, dtype='int32').reshape(-1,1)

print(feature_matrix_X.shape, label_Y.shape)
print(feature_matrix_X[:5])
print(label_Y[:5])
```
